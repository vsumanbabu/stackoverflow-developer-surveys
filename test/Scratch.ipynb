```python
#f_db_desired_list = lambda x: x['DatabaseDesireNextYear'].str.split(";")

df['SplitDatabaseWorkedWith']=df['DatabaseWorkedWith'].str.split(";")
df['SplitDatabaseDesireNextYear']=df['DatabaseDesireNextYear'].str.split(";")

df1=df['SplitDatabaseWorkedWith'].apply(pd.Series).stack().reset_index(drop=True)
df2=df['SplitDatabaseDesireNextYear'].apply(pd.Series).stack().reset_index(drop=True)
```

```python
def get_df(t,cs,cd):
    
    
    dft=t
    dft


    dft['c1']=dft[cs].str.split(";")
    dft1=dft['c1'].apply(pd.Series).stack().reset_index(drop=True)


    dft11=dft1.to_frame()
    dft11.columns = [cd]
    dft11['cntr']=1
    dft11


    dft111=dft11.groupby(cd).sum()['cntr'].to_frame()
    dft111.columns = ['Counts']
    dft111 = dft111.sort_values('Counts',ascending=False)
    dft111
    
    
    return dft111
```

```python
t=[{'c0':'DynamoDB;MariaDB;Microsoft SQL Server;MySQL'},
    {'c0':'MariaDB;MySQL;Oracle;SQLite'}
      ]
pt=pd.DataFrame(t)
dd=get_df(pt,'c0','c2')
dd
```
